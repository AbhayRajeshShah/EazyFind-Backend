```python
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
cols = df.columns
```

```python
df.dropna(subset=cols, inplace=True)
```

```python
df
```

```python
restaurants = (
    df[['restaurant_name', 'url', 'city', 'area', 'cost_for_two', 'rating', 'page']]
    .drop_duplicates()
    .reset_index(drop=True)
)

restaurants['restaurant_id'] = restaurants.index + 1

restaurants = restaurants[
    ['restaurant_id', 'restaurant_name', 'url', 'city', 'area',
     'cost_for_two', 'rating', 'page']
]

restaurants['cost_for_two'] = (
    restaurants['cost_for_two']
    .str.extract(r'(\d+)', expand=False)
    .astype('Int64')
)
```

```python
restaurants
```

```python
meal_types = (
    df['meal_type']
    .dropna()
    .str.lower()
    .str.split(',')
    .explode()
    .str.strip()
    .drop_duplicates()
    .reset_index(drop=True)
    .to_frame(name='meal_type')
)

meal_types['meal_type_id'] = meal_types.index + 1
meal_types = meal_types[['meal_type_id', 'meal_type']]
```

```python
meal_types
```

```python
restaurant_meal_types = (
    df[['restaurant_name', 'url', 'meal_type']]
    .dropna(subset=['meal_type'])
    .assign(meal_type=lambda x: x['meal_type']
            .str.lower().str.split(','))
    .explode('meal_type')
)

restaurant_meal_types['meal_type'] = restaurant_meal_types['meal_type'].str.strip()

restaurant_meal_types = restaurant_meal_types.merge(
    restaurants[['restaurant_id', 'restaurant_name', 'url']],
    on=['restaurant_name', 'url']
)

restaurant_meal_types = restaurant_meal_types.merge(
    meal_types,
    on='meal_type'
)

restaurant_meal_types = restaurant_meal_types[
    ['restaurant_id', 'meal_type_id']
].drop_duplicates()
```

```python
restaurant_meal_types
```

```python
cuisines = (
    df['cuisine']
    .dropna()
    .str.lower()
    .str.split(',')
    .explode()
    .str.strip()
    .drop_duplicates()
    .reset_index(drop=True)
    .to_frame(name='cuisine')
)

cuisines['cuisine_id'] = cuisines.index + 1
cuisines = cuisines[['cuisine_id', 'cuisine']]
```

```python
restaurant_cuisines = (
    df[['restaurant_name', 'url', 'cuisine']]
    .dropna(subset=['cuisine'])
    .assign(cuisine=lambda x: x['cuisine']
            .str.lower().str.split(','))
    .explode('cuisine')
)

restaurant_cuisines['cuisine'] = restaurant_cuisines['cuisine'].str.strip()

restaurant_cuisines = restaurant_cuisines.merge(
    restaurants[['restaurant_id', 'restaurant_name', 'url']],
    on=['restaurant_name', 'url']
)

restaurant_cuisines = restaurant_cuisines.merge(
    cuisines,
    on='cuisine'
)

restaurant_cuisines = restaurant_cuisines[
    ['restaurant_id', 'cuisine_id']
].drop_duplicates()
```

```python
restaurant_cuisines
```

```python
restaurants = (
    df[['restaurant_name', 'url', 'city', 'area', 'cost_for_two', 'rating', 'page','offer']]
    .drop_duplicates()
    .reset_index(drop=True)
)

restaurants['restaurant_id'] = restaurants.index + 1

restaurants = restaurants[
    ['restaurant_id', 'restaurant_name', 'url', 'city', 'area',
     'cost_for_two', 'rating', 'page','offer']
]

restaurants['cost_for_two'] = (
    restaurants['cost_for_two']
    .str.extract(r'(\d+)', expand=False)
    .astype('Int64')
)
```

```python
restaurants['offer'].unique()
```

```python
restaurants['percentages'] = df['offer'].str.findall(r'(\d+)\s*%').apply(
    lambda x: [int(p)/100 for p in x]
)
```

```python
import math

def effective_discount(x):
    # No discount
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return 0.0

    # Single percentage stored as float
    if isinstance(x, (int, float)):
        return float(x)

    # Proper list of percentages
    remaining = 1.0
    for d in x:
        remaining *= (1 - d)
    return round(1 - remaining, 4)
```

```python
restaurants['effective_discount'] = restaurants['percentages'].apply(effective_discount)
```
